**Loading the dataset**

```python
data = "D:/Jayanth/Desktop/Data"
dataset = ImageFolder(data, transform=ToTensor())
```

```python
print('Classes in the dataset  :',dataset.classes)
print('Total number of images in the dataset  :' + str(len(dataset)))
```

**Find the number of images in each class**

```python
classes = dataset.classes
num_classes =len(dataset.classes)
img_dict = {}
for i in range(num_classes):
    img_dict[classes[i]] = 0
    
for i in range(len(dataset)):
    img,label = dataset[i]
    img_dict[classes[label]] +=1
```
